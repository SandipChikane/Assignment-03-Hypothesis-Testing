# **Question - 01**

```python
# A&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. 
# A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 
# 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.
```

```python
from google.colab import files
upload = files.upload()
```

```python
data=pd.read_csv('Cutlets.csv')
data.head()
```

```python
#Lets define Hypothesis
# Null hypothesis H0: Mean_A = Mean_B       Diameters are equal to each other
# Null Hypothesis H1: Mean_A!= Mean_B       Diameters are not equal to each other

F = stats.f_oneway(data['Unit A'],data['Unit B'])
# For 5% level of significance 

if(F.pvalue>0.05):
  print('We fail to reject null hypothesis.Hence,there is no significant difference in the diameter of the cutlet between two units')
else:
  print('We reject the null hypothesis. Hence,there is significant difference in the diameter of the cutlet between two units ')
```

```python
data1 = pd.read_csv('LabTAT.csv')
data1.head()
```

```python
#    Lets define the Hypothesis
#    Null hypothesis H0: All laboratories average are equal.
#    Alternative Hypothesis  H1:  There are at least one laboratory average that are significantly different from each other.

#  Calculate the p-value for one way anova.
p_value =(stats.f_oneway(data1['Laboratory 1'],data1['Laboratory 2'],data1['Laboratory 3'],data1['Laboratory 4'])).pvalue

if (p_value>0.05):
  print('We fail to reject null hypothesis. Hence,there is no any difference in average TAT among the different laboratories')
else:
  print('We rject the null hypothesis.Hence, there is difference in average TAT among the different laboratories')
```
